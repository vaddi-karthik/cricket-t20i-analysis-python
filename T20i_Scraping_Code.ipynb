```python
import time
import requests
import pandas as pd
from bs4 import BeautifulSoup

all_rows = []

for page in range(1, 40):
    print(f"Scraping page {page}/40 ...")

    url = f"https://stats.espncricinfo.com/ci/engine/stats/index.html?class=11;page={page};template=results;type=batting;size=200"
    headers = {"User-Agent": "Mozilla/5.0"}

    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")

    # Find the correct table — the one that has "Player" in its header
    table = None
    for t in soup.find_all("table", class_="engineTable"):
        if t.find("tr", class_="headlinks"):
            headers_row = t.find("tr", class_="headlinks")
            if "Player" in headers_row.text:
                table = t
                break

    if table is None:
        print(f"  -> Table not found on page {page}, skipping.")
        continue

    rows = table.find_all("tr", class_="data1")

    for row in rows:
        cells = row.find_all("td")
        if len(cells) < 11:
            continue
        all_rows.append({
            "Player": cells[0].text.strip(),
            "Span"  : cells[1].text.strip(),
            "Mat"   : cells[2].text.strip(),
            "Inns"  : cells[3].text.strip(),
            "NO"    : cells[4].text.strip(),
            "Runs"  : cells[5].text.strip(),
            "HS"    : cells[6].text.strip(),
            "Ave"   : cells[7].text.strip(),
            "100"   : cells[8].text.strip(),
            "50"    : cells[9].text.strip(),
            "0"     : cells[10].text.strip(),
        })

    time.sleep(1)

df = pd.DataFrame(all_rows)
df.to_csv("t20i_batting_stats.csv", index=False)
print(f"Done! {len(df)} rows saved.")
```
